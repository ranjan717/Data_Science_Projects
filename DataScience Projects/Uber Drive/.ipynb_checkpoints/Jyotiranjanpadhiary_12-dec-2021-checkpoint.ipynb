## Q17. What is the most frequent route taken by Uber drivers. (3 points)
#### Note: This question is based on the new dataframe with no 'na' values.
#### Hint-Print the most frequent route taken by Uber drivers (Route= combination of START & END points present in the Data set).

```python
dfx=df.groupby(['START*','STOP*']).size().reset_index().rename(columns={0:'count'}) #a dataframe of start stop destination combination count is displayed
dfx
```

```python
dfx['count'].max() #from the dataframe max count of combination found is 52
```

```python
dfx[dfx['count']==52] # for max count 52 the start and stop point is displayed in the dataframe
```

```python
#the most frequent route taken by uber drivers is from cary to Morrisville
```

## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
plt.figure(figsize=(15,6))
sns.barplot(x='MILES*',y='PURPOSE*',estimator=np.sum,data=uber,ci=None) # a barplot is displayed 
plt.show()
```

```python
#people have travelled more miles for the purpose of meeting and the least miles travelled is for airport purpose.
```

## Q22. What percentage of Miles were clocked under Business Category and what percentage of Miles were clocked under Personal Category ? (3 points)

### Note:Use the original dataframe without dropping the 'NA' values. 


```python
uber[uber['CATEGORY*']=='Business']['MILES*'].sum()/uber['MILES*'].sum()*100 # percentage of business category miles covered
```

```python
uber[uber['CATEGORY*']=='Personal']['MILES*'].sum()/uber['MILES*'].sum()*100 #percentage of personal category miles covered
```

```python
#94.11% of miles were clocked under Business category and 5.88% of miles clocked under Personal category
```
